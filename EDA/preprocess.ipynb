- we need to optimize this dataframe in order to load it fully into the RAM 
- for that i will try to minimize the object-typed colunms , which are mainly strings here 
- i will be converting (encoding) them from strings to numbers 
because strings use up to 10 times more the storage when loading with pandas , because they are treated as python objects , unlike numbers which are treated as numpy arrays 

```python
import numpy as np
#cyclical encoding of days 

mp={
    'Saturday':5,
    'Sunday':6, 
    'Monday':0,
    'Tuesday':1,
    'Wednesday':2,
    'Thursday':3,
    'Friday':4,
    
}
frst=True 
i=1
for chk in pd.read_csv('final3.csv',chunksize=1_000_000): 
    
    print(f'{i*1000000} chunks processed')
    i+=1 

    chk['weekday']=chk['weekday'].map(mp)
    chk['weekday_sin']=np.sin(2*np.pi*(chk['weekday']/7))
    chk['weekday_cos']=np.cos(2*np.pi*(chk['weekday']/7))

    chk.drop('weekday',inplace=True,axis=1)
    chk.to_csv('final4.csv',index=False,mode='a',header=frst)
    frst=False
    
    del chk 
```

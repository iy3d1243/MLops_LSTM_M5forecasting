```python
# Requires: pip install pympler
from pympler import asizeof
import pandas as pd
import numpy as np
import sys, gc, types

def sizeof_obj(obj):
    """Return best-effort deep size in bytes for common types."""
    try:
        # fast path for pandas and numpy
        if isinstance(obj, pd.DataFrame) or isinstance(obj, pd.Series):
            # pandas reports memory for its internals reliably
            try:
                return obj.memory_usage(deep=True).sum()
            except Exception:
                return asizeof.asizeof(obj)
        if isinstance(obj, np.ndarray):
            # numpy nbytes is exact for array data
            try:
                return obj.nbytes
            except Exception:
                return asizeof.asizeof(obj)
        # fallback to pympler deep size (handles nested structures)
        return asizeof.asizeof(obj)
    except Exception:
        # last fallback: sys.getsizeof (shallow)
        try:
            return sys.getsizeof(obj)
        except Exception:
            return 0

def human_readable(nbytes):
    for unit in ['B','KB','MB','GB','TB']:
        if nbytes < 1024.0:
            return f"{nbytes:3.2f} {unit}"
        nbytes /= 1024.0
    return f"{nbytes:.2f} PB"

def list_vars_sizes(scope=None, top=30, exclude_private=True):
    """
    List variables and their sizes in the provided scope (defaults to globals()).
    Returns a sorted list of tuples: (name, type, size_bytes).
    """
    if scope is None:
        # inspect caller globals if running in notebook top-level, otherwise globals()
        scope = globals()
    items = []
    for name, val in scope.items():
        if exclude_private and name.startswith('_'):
            continue
        # skip modules and functions to avoid noise (but you can include them if you want)
        if isinstance(val, types.ModuleType) or isinstance(val, types.FunctionType):
            continue
        try:
            size = sizeof_obj(val)
            items.append((name, type(val).__name__, size))
        except Exception:
            items.append((name, type(val).__name__, 0))
    items_sorted = sorted(items, key=lambda x: x[2], reverse=True)
    # print nicely
    print(f"{'name':30} {'type':20} {'size':12}")
    print("-"*66)
    for name, typ, size in items_sorted[:top]:
        print(f"{name:30} {typ:20} {human_readable(size):12}")
    return items_sorted

def delete_var(name, scope=None):
    """Delete variable by name from scope (globals by default), then gc.collect()."""
    if scope is None:
        scope = globals()
    if name in scope:
        try:
            del scope[name]
            gc.collect()
            print(f"Deleted `{name}` and ran gc.collect()")
            return True
        except Exception as e:
            print(f"Failed to delete `{name}`: {e}")
            return False
    else:
        print(f"No variable named `{name}` found in scope.")
        return False

# Example usage:
# 1) run list_vars_sizes() to see big objects
# 2) pick a variable name printed there and call delete_var('that_name')
#
# Run it now:
if __name__ == "__main__":
    list_vars_sizes()
```
